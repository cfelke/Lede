### Loop through each result and print each person's name

You'll get an error because the first one doesn't have a name. How do you make that not happen?! If you want to ignore an error, you use code like this:

```python

```

It should help you out. If you don't want to print anything, you can type `pass` instead of the `print` statement.

**Why doesn't the first one have a name?**

````python
cosmetologists = driver.find_elements_by_tag_name('tr')
for name in cosmetologists[:10]:
    try:
        names_cosmetologists = name.find_element_by_class_name('results_text')
        print(names_cosmetologists.text)
    except:
        print('It did not work')
````

## Saving the results

### Loop through each result to create a list of dictionaries

Each dictionary must contain

- Person's name
- Violation description
- Violation number
- License Numbers
- Zip Code
- County
- City

Create a new dictionary for each result (except the header).

> *Tip: If you want to ask for the "next sibling," you can't use `find_next_sibling` in Selenium, you need to use `element.find_element_by_xpath("following-sibling::div")` to find the next div, or `element.find_element_by_xpath("following-sibling::*")` to find the next anything.

````python
rows = driver.find_elements_by_tag_name('tr')
cosmetologists = []

for row in rows[1:11]:
    nguyens = {}
    name = row.find_element_by_class_name('results_text')
    nguyens['name'] = name.text
    
    violation_descr = row.find_elements_by_tag_name('td')
    nguyens['violation_descr'] = violation_descr[2].text
    
    violation_num = row.find_elements_by_class_name('results_text')
    nguyens['violation_num']= violation_num[-2].text
    
    license_num = row.find_elements_by_class_name('results_text')
    nguyens['license_num'] = license_num[-3].text
    
    zip_code = row.find_elements_by_class_name('results_text')
    nguyens['zip_code'] = license_num[-4].text
    
    county = row.find_elements_by_class_name('results_text')
    nguyens['county'] = county[-5].text
    
    city = row.find_elements_by_class_name('results_text')
    nguyens['city'] = city[-6].text
    
    cosmetologists.append(nguyens)
  
cosmetologists
    
````
